```python
import yfinance as yf

# Define the tickers and time period
tickers = ["BARC.L", "HSBA.L"]
start_date = "2021-10-01"
end_date = "2023-10-01"

# Download historical data
data = yf.download(tickers, start=start_date, end=end_date)

# Display the first few rows of the data
print(data.head())
```

```python
# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Barclays on the primary axis
ax1.plot(data['BARC.L']['Close'], label='Barclays (BARC.L)', color='blue')
ax1.set_xlabel("Date")
ax1.set_ylabel("Barclays Price (GBP)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary axis for HSBC
ax2 = ax1.twinx()
ax2.plot(data['HSBA.L']['Close'], label='HSBC (HSBA.L)', color='orange')
ax2.set_ylabel("HSBC Price (GBP)", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add title and legends
plt.title("Barclays and HSBC Closing Prices (2023)")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.grid()
plt.show()
```

```python
import numpy as np

# Calculate daily returns
barclays_returns = data['BARC.L']['Close'].pct_change().dropna()
hsbc_returns = data['HSBA.L']['Close'].pct_change().dropna()

# Calculate correlation
correlation = np.corrcoef(barclays_returns, hsbc_returns)[0, 1]
print(f"Correlation between Barclays and HSBC: {correlation:.4f}")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# Download data for Barclays and HSBC
tickers = ["BARC.L", "HSBA.L"]
data = yf.download(tickers, start="2023-01-01", end="2023-10-01", group_by='ticker')

# Function to calculate SMAs and generate positions
def sma_strategy(data, ticker, short_window=20, long_window=50):
    # Calculate SMAs
    data[ticker, 'SMA1'] = data[ticker]['Close'].rolling(window=short_window).mean()
    data[ticker, 'SMA2'] = data[ticker]['Close'].rolling(window=long_window).mean()
    
    # Generate positions (1 for buy, -1 for sell)
    data[ticker, 'positions'] = np.where(data[ticker, 'SMA1'] > data[ticker, 'SMA2'], 1, -1)
    
    # Drop rows with NaN values (due to SMA calculations)
    data.dropna(inplace=True)
    
    # Plot the data
    ax = data[ticker][['Close', 'SMA1', 'SMA2', 'positions']].plot(
        figsize=(10, 6), secondary_y='positions')
    ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
    plt.title(f"{ticker} SMA Crossover Strategy (2023)")
    plt.xlabel("Date")
    plt.ylabel("Price (GBP)")
    plt.grid()
    plt.show()

# Apply strategy to Barclays and HSBC
sma_strategy(data, 'BARC.L')
sma_strategy(data, 'HSBA.L')
```
